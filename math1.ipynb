3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
    1. окружности,
    2. эллипса,
    3. гиперболы.

```python
# окружность
R = 2
x0, y0 = (1, 2) # координаты центра окружности
x = np.linspace(x0-R, x0+R, 201)
y1 = [] # правая половина
y2 = [] # левая половина
for x_ in x:
    y1.append(np.sqrt(R**2 - (x_-x0)**2) + y0)
    y2.append(-np.sqrt(R**2 - (x_-x0)**2) + y0)
plt.plot(x, y1, c='red')
plt.plot(x, y2, c='red')
plt.gca().set_aspect('equal')
plt.grid()
```

```python
# эллипс
a = 2
b = 1
x0, y0 = (1, 2) # координаты центра
x = np.linspace(x0-a, x0+a, 201)
y1 = [] # правая половина
y2 = [] # левая половина
for x_ in x:
    y1.append(np.sqrt(b**2 - b**2/a**2 * (x_-x0)**2) + y0)
    y2.append(-np.sqrt(b**2 - b**2/a**2 * (x_-x0)**2) + y0)
plt.plot(x, y1, c='red')
plt.plot(x, y2, c='red')
plt.gca().set_aspect('equal')
plt.grid()
```

```python
# гипербола
a = 2
b = 1
x0, y0 = (1, 2) # координаты центра
x1 = np.linspace(x0-3*a, x0-a, 201)
x2 = np.linspace(x0+a, x0+3*a, 201)
y11 = [] 
y12 = [] 
y21 = [] 
y22 = [] 
for x_ in x1:
    y11.append(np.sqrt(-b**2 + b**2/a**2 * (x_-x0)**2) + y0)
    y12.append(-np.sqrt(-b**2 + b**2/a**2 * (x_-x0)**2) + y0)
    
for x_ in x2:    
    y21.append(np.sqrt(-b**2 + b**2/a**2 * (x_-x0)**2) + y0)
    y22.append(-np.sqrt(-b**2 + b**2/a**2 * (x_-x0)**2) + y0)
plt.plot(x1, y11, c='red')
plt.plot(x1, y12, c='red')
plt.plot(x2, y21, c='red')
plt.plot(x2, y22, c='red')
#plt.gca().set_aspect('equal', adjustable='box')
plt.grid()
```

```python
# трехмерный график эллиптический парабалоид
fig = figure()
ax = Axes3D(fig)
x = np.arange(-5, 5, 0.5)
y = np.arange(-5, 5, 0.5)
x, y = np.meshgrid(x, y)
z = x**2/4 + y**2/9
ax.plot_wireframe(x, y, z, color='r')
plt.gca().set_aspect('equal')
show()
```

```python
# трехмерный график сфера
fig = figure()
ax = Axes3D(fig)
u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
x = np.cos(u)*np.sin(v)
y = np.sin(u)*np.sin(v)
z = np.cos(v)
ax.plot_wireframe(x, y, z, color="r")
plt.gca().set_aspect('equal')
show()
```

1) Решите систему уравнений:

y = x^2 – 1

exp(x) + x*(1 – y) = 1



```python
# для наглядности нарисуем графики
from scipy.optimize import fsolve

x0 = np.linspace(-2.5, 4.5, 41)
y1 = x0**2 - 1
y2 = 1 - (1-np.exp(x0))/x0

def equations(p):
    x, y = p
    return (x**2 - 1 - y, np.exp(x) + x*(1-y) - 1)


x_solv = []
y_solv = []
x, y = fsolve(equations, (2.75, 8))
x_solv.append(x)
y_solv.append(y)
print(x, y)
x, y = fsolve(equations, (4.25, 18))
x_solv.append(x)
y_solv.append(y)
print(x, y)
x, y = fsolve(equations, (-2, 2))
x_solv.append(x)
y_solv.append(y)
print(x, y)

plt.scatter(x_solv, y_solv, color='blue', label='roots')
plt.plot(x0, y1, c='r', label='x**2 - 1')
plt.plot(x0, y2, c='g', label='1 - (1-np.exp(x))/x')
plt.legend()
plt.grid()
```
